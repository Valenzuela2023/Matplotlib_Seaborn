#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.linspace(0, 5, 100) 

y1 = x**2
y2 = np.exp(x)

# Create a figure and subplot
fig, ax = plt.subplots()

# Plot y = x^2 with normal scale
ax.plot(x, y1, label='y = x^2', color='blue')

# Plot y = e^x with normal scale
ax.plot(x, y2, label='y = e^x', color='green')

# Set the scale for x and y axes to logarithmic
ax.set_xscale('log')
ax.set_yscale('log')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Convert the 'Date' column to datetime
fitbit['Date'] = pd.to_datetime(fitbit['Date'])

# Extract the day of the week (0=Monday, 1=Tuesday, etc.) and store it in 'Day_of_week'
fitbit['Day_of_week'] = fitbit['Date'].dt.dayofweek

# Set the figure size to 12x4 inches
plt.figure(figsize=(12, 4))

# subplots
plt.subplot(131)  # 1 row, 3 columns, 1st plot
monthly_steps = fitbit.groupby('Months_encoded')['Steps'].mean()
monthly_steps.plot(kind='bar', color='blue', alpha=0.7)
plt.xlabel('Months')
plt.ylabel('Average Steps')
plt.title('Average Steps by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.subplot(132)  # 1 row, 3 columns, 2nd plot
fitbit['Work_or_Weekend'] = fitbit['Work_or_Weekend'].map({0: 'Weekend', 1: 'Workday'})
workday_weekend_steps = fitbit.groupby('Work_or_Weekend')['Steps'].mean()
workday_weekend_steps.plot(kind='bar', color='green', alpha=0.7)
plt.xlabel('Workday or Weekend')
plt.ylabel('Average Steps')
plt.title('Average Steps by Workday/Weekend')

plt.subplot(133)  # 1 row, 3 columns, 3rd plot
day_of_week_steps = fitbit.groupby('Day_of_week')['Steps'].mean()
day_of_week_steps.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
day_of_week_steps.plot(kind='bar', color='orange', alpha=0.7)
plt.xlabel('Day of the Week')
plt.ylabel('Average Steps')
plt.title('Average Steps by Day of the Week')

# Create a 'figures' folder in your current working directory if it doesn't exist
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

# Save figures in the 'figures' folder
plt.savefig('figures/fitbit_visualizations.png', bbox_inches='tight')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
feature_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')]

# subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Loop through the feature pairs and plot scatter plots
for i, (x_feature, y_feature) in enumerate(feature_pairs):
    ax = axes[i]
    ax.scatter(fitbit[x_feature], fitbit[y_feature], alpha=0.5)
    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)
    ax.set_title(f'Scatter Plot: {x_feature} vs. {y_feature}')

# spacing between subplots
plt.tight_layout()
plt.show()
```
